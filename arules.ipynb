```python
from ast import literal_eval
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
posts = pd.read_csv("./data/refined/post_with_seg.csv")
comments_by_author = pd.read_csv("./data/refined/comments_by_author_with_seg.csv")
posts['word_list'] = posts['word_list'].apply(literal_eval)
comments_by_author['word_list'] = comments_by_author['word_list'].apply(literal_eval)
comments_by_author['fanpage'] = comments_by_author['post_url'].map(posts.set_index('post_url')['fanpage'])
comments_by_author['phase'] = comments_by_author['post_url'].map(posts.set_index('post_url')['phase'])
```

## 貼文

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(posts['word_list']).transform(posts['word_list'])
post_df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules

post_frequent_itemsets = fpgrowth(post_df, min_support=0.02, use_colnames=True)

post_arules = association_rules(post_frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
post_arules['antecedents'] = post_arules['antecedents'].apply(list).apply(lambda x:"+".join(x))
post_arules['consequents'] = post_arules['consequents'].apply(list).apply(lambda x:"+".join(x))
```

```python
post_arules
```

# 貼文- 留言

```python
a = comments_by_author.set_index('post_url')['word_list'].explode().dropna()
a = a[a.apply(len)>1]
b = a.groupby('post_url').apply(list).reset_index()
```

```python
posts_refined = posts
posts_refined['word_list'] = posts_refined['word_list'].apply(lambda x:[i for i in x if len(i)>1 and i not in ['三倍券','振興券']])
```

```python
post_and_comment = pd.merge(
    posts_refined[['post_url', 'word_list']], 
    b[['post_url','word_list']], 
    on='post_url', 
    suffixes=('_post','_comment')
)
```

```python
for row in post_and_comment.loc[post_and_comment['word_list_comment'].isnull(), 'word_list_comment'].index:
    post_and_comment.at[row, 'word_list_comment'] = []
```

```python
post_and_comment['word_list_post'] = post_and_comment['word_list_post'].apply(lambda x:['post_'+i for i in x])
post_and_comment['word_list_comment'] = post_and_comment['word_list_comment'].apply(lambda x:['comment_'+str(i) for i in x])
```

```python
post_and_comment['word_list'] = post_and_comment['word_list_post'] + post_and_comment['word_list_comment']
```

```python
te = TransactionEncoder()
encoded_postcomment = te.fit(post_and_comment['word_list']).transform(post_and_comment['word_list'])
postandcomment_df = pd.DataFrame(encoded_postcomment, columns=te.columns_)
```

```python
drop_cols = []
for col in postandcomment_df.columns.values:
    if postandcomment_df[col].sum() < 10 or any([x in col for x in ["https","PauseGIF","Pause", "media1tenorco"]]):
        drop_cols.append(col)
```

```python
postandcomment_df.drop(drop_cols,axis='columns',inplace=True)
```

```python
comment_cols = [col for col in postandcomment_df.columns.values if col.startswith('comment_')]
post_cols = [col for col in postandcomment_df.columns.values if col.startswith('post_')]
```

```python
len(comment_cols)*len(post_cols)
```

```python
%%time
from itertools import product
min_supp = 10
freq = pd.DataFrame()
freq['itemsets'] = [(comment_col, post_col) for comment_col, post_col in product(comment_cols, post_cols)]
freq['support'] = freq['itemsets'].apply(lambda x:(postandcomment_df[x[0]] & postandcomment_df[x[1]]).sum())
freq = freq[freq['support']>min_supp]
```

```python
freq.shape
```

```python
atomic = pd.DataFrame()
atomic['itemsets'] = [(col,) for col in postandcomment_df.columns.values]
atomic['support'] = atomic['itemsets'].apply(lambda x:postandcomment_df[x[0]].sum())
```

```python
postandcomment_frequent_itemsets = pd.concat([freq,atomic],axis='rows')
```

```python
postandcomment_frequent_itemsets['support'] = postandcomment_frequent_itemsets['support']/postandcomment_df.shape[0]
```

```python
# postandcomment_frequent_itemsets = fpgrowth(postandcomment_df, min_support=0.4, use_colnames=True, max_len=2)
```

```python
postandcomment_arules = association_rules(postandcomment_frequent_itemsets, metric="confidence", min_threshold=0.9)
```

```python
postandcomment_arules['antecedents'] = postandcomment_arules['antecedents'].apply(list).apply(lambda x:"+".join(x))
postandcomment_arules['consequents'] = postandcomment_arules['consequents'].apply(list).apply(lambda x:"+".join(x))
```
