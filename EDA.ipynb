```python
import pandas as pd
pd.set_option('display.max_rows', 1000)
1
```

```python
posts = pd.read_csv("./data/refined/posts_v1.csv")
comments = pd.read_csv("./data/refined/comments.csv")
```

```python
posts.head()
```

```python
posts['post_time'] = pd.to_datetime(posts['post_time'])
```

```python
first = posts[(posts['post_time']>pd.to_datetime('2020-01-31'))&(posts['post_time']<pd.to_datetime('2020-06-02'))]
second = posts[(posts['post_time']>=pd.to_datetime('2020-06-02'))&(posts['post_time']<pd.to_datetime('2020-07-15'))]
third = posts[(posts['post_time']>=pd.to_datetime('2020-07-15'))&(posts['post_time']<pd.to_datetime('2021-01-01'))]
# forth = posts[(posts['post_time']>=pd.to_datetime('2021-01-01'))&(posts['post_time']<pd.to_datetime('2021-02-28'))]
```

```python
first_comments = comments[comments['post_url'].isin(first['post_url'].unique())]
second_comments = comments[comments['post_url'].isin(second['post_url'].unique())]
third_comments = comments[comments['post_url'].isin(third['post_url'].unique())]
# forth_comments = comments[comments['post_url'].isin(forth['post_url'].unique())]
```

```python
third_comments.head()
```

```python
phase_post = [first, second, third]
phase_comment = [first_comments,second_comments,third_comments]
```

```python
metrics = pd.DataFrame(index=['first','second','third'])
metrics['post_cnt'] = [len(i) for i in phase_post]
metrics['avg_reaction'] = [i['total_reaction'].mean() for i in phase_post]
metrics['median_reaction'] = [i['total_reaction'].median() for i in phase_post]
metrics['avg_reaction_angry'] = [i['Angry'].mean() for i in phase_post]
metrics['avg_reaction_like'] = [i['Like'].mean() for i in phase_post]
metrics['avg_like_ratio'] = [(i['Like']/i['total_reaction']).mean() for i in phase_post]
metrics['comment_cnt'] = [len(i) for i in phase_comment]
metrics['avg_comment_cnt_per_post'] = [i.groupby('post_url')['comment'].count().mean() for i in phase_comment]
metrics['std_comment_cnt_per_post'] = [i.groupby('post_url')['comment'].count().std() for i in phase_comment]
metrics['median_comment_cnt_per_post'] = [i.groupby('post_url')['comment'].count().median() for i in phase_comment]
```
